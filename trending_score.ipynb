```python
from cmath import nan
like = results['likes'].tolist()
retweet = results['retweets'].tolist()
reply = results['replies'].tolist()
def getUniqueWords(allWords) :
    uniqueWords = [] 
    for i in allWords:
        if not i in uniqueWords:
            uniqueWords.append(i)
    return uniqueWords

def like_count(df, category):
    cat = df[category].tolist()
    sat = getUniqueWords(cat)
    count=0
    like_dict ={}
    for i in range(0, len(sat), 1):
        for j in range(0, len(like), 1):
            if((sat[i] == cat[j])&(like[j]>=0)):
                count+=like[j]
        #print(sat[i],':',count)
        if(count!=0):
            like_dict[sat[i]] = count
        count=0
    return (like_dict)

def reply_count(df, category):
    cat = df[category].tolist()
    sat = getUniqueWords(cat)
    count=0
    like_dict ={}
    for i in range(0, len(sat), 1):
        for j in range(0, len(reply), 1):
            if((sat[i] == cat[j])&(reply[j]>2)):
                count+=like[j]
        #print(sat[i],':',count)
        if(count!=0):
            like_dict[sat[i]] = count
        count=0
    return (like_dict)

def retweet_count(df, category):
    cat = df[category].tolist()
    sat = getUniqueWords(cat)
    count=0
    like_dict ={}
    for i in range(0, len(sat), 1):
        for j in range(0, len(retweet), 1):
            if((sat[i] == cat[j])&(retweet[j]>2)):
                count+=retweet[j]
        #print(sat[i],':',count)
        if(count!=0):
            like_dict[sat[i]] = count
        count=0
    return (like_dict)
```

```python
suml = (sum(like_articleType.values()))
sumr = (sum(retweet_articleType.values()))
sumre = (sum(reply_articleType.values()))

sumlu = (sum(like_usage.values()))
sumru = (sum(retweet_usage.values()))
sumreu = (sum(reply_usage.values()))

sumlp = (sum(like_patternType.values()))
sumrp = (sum(retweet_patternType.values()))
sumrep = (sum(reply_patternType.values()))

sum_all = suml+sumr+sumre
sum_allu = sumlu+sumru+sumreu
sum_allu = sumlu+sumru+sumreu

for k, v in retweet_articleType.items():
    retweet_articleType[k] = (v/sumr)*100
for k, v in retweet_usage.items():
    retweet_usage[k] = (v/sumru)*100
for k, v in retweet_patternType.items():
    retweet_patternType[k] = (v/sumrp)*100

for k, v in like_articleType.items():
    like_articleType[k] = (v/suml)*100
for k, v in like_usage.items():
    like_usage[k] = (v/sumlu)*100
for k, v in like_patternType.items():
    like_patternType[k] = (v/sumlp)*100
    
for k, v in reply_articleType.items():
    reply_articleType[k] = (v/sumre)*100
for k, v in reply_usage.items():
    reply_usage[k] = (v/sumreu)*100
for k, v in reply_patternType.items():
    reply_patternType[k] = (v/sumrep)*100
```

```python
import itertools
import collections
# using defaultdict
article = collections.defaultdict(int)
usage = collections.defaultdict(int)
pattern = collections.defaultdict(int)
# iterating key, val with chain()
for key, val in itertools.chain(like_articleType.items(), retweet_articleType.items(), reply_articleType.items()):
    article[key] += val 
    
for key, val in itertools.chain(like_usage.items(), retweet_usage.items(), reply_usage.items()):
    usage[key] += val 

for key, val in itertools.chain(like_patternType.items(), retweet_patternType.items(), reply_patternType.items()):
    pattern[key] += val 
```
